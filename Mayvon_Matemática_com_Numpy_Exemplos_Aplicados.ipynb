**Autovalores e Autovetores**

Autovetores e Autovalores (Eigenvectors, Eigenvalues)

```python
B = np.arange(2,8,2)
print(B)
```

```python
# Encontrar os autovalores e autovetores para uma matriz quadrada simples
A = np.diag(np.arange(2,8,2)) #preenchendo uma matriz só na diagonal, no intervalo de 2 a 6, considerando de 2 em 2.
A
```

```python
# Retorna os autovalores e autovetores para a Matriz A
autovalores, autovetores = np.linalg.eig(A)

print("Os autovalores são:", autovalores)
print("Os autovetores são:\n", autovetores)
```

**Visualização de Gradientes com Matplotlib**

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,10)
y = np.linspace(0,10,10)
xv, yv = np.meshgrid(x,y)

def f(x,y):
    return  2 * (x**2) * y + (2*x) #função

z = f(xv, yv)
Gy,Gx = np.gradient(f(xv, yv)) #np.gradiente

plt.figure(figsize=(8,5))
plt.pcolor(xv, yv, z, cmap='plasma')
plt.colorbar()
plt.quiver(xv, yv, Gx, Gy, scale = 3000, color = 'w')
plt.title('Gradiente de f(x,y) = 2x^2 * y + 2x')
plt.show()
```

Encontre o mínimo global para a função objetiva, $f (x) = 2x^2 - 3$. OBS: Utilize o intervalo $ 0 <= x <= 10$

```python
def f(x):
    return 2*(x**2) - 3

x = np.linspace(1,10,10)

print(x)

x = f(x)

print("Após a primeira derivação:", np.diff(x, n=1))
print("Após a segunda derivação:", np.diff(x, n=2))
print("Após a terceira derivação:", np.diff(x, n=3))
```

**Matrizes de Covariância**

A matriz de covariância busca estabelecer alguma relação (ou ausência de relação) entre as variáveis de um determinado conjunto de dados. Muitas variáveis podem estar altamente relacionadas a ponto de gerarem redundância nos dados.


*   Se positivo, então as duas variáveis crescem ou decrescem juntas (correlacionadas);
*   Se negativo, então uma das variáveis cresce enquanto a outra decresce (inversamente correlacionadas).

```python
# Covariância - busca estabelecer alguma relação entre as variáveis

import numpy as np

x = np.array([[0, 1, 2], [2, 1, 0]])

print("Matriz x:\n",x)
print('-------------------------------------')
print("Matriz de covariância para x:\n", np.cov(x))
```
